**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# our labels are on columns instead of row, use transpose to flip dataframe
properties_T = properties.transpose()

# show resultant df
properties_T.head()
```

```python
# reset index to have numbers instead of borough names
properties_T = properties_T.reset_index()
# set first row to be the column labels
properties_T.columns = properties_T.iloc[0]
# remove redundant first row
properties_T = properties_T.drop(0)
properties_T.head(10)
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
#rename columns Unnamed and Nan
properties_T = properties_T.rename(columns = {'Unnamed: 0':'London_Borough', pd.NaT: 'ID'})
properties_T.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# melt muliple columns to one date column for Month
melt_properties = pd.melt(properties_T, id_vars= ['London_Borough', 'ID'])
# rename the date column from 0 to Month and the value column to Average_price
melt_properties = melt_properties.rename(columns = {0: 'Month', 'value': 'Average_price'})
melt_properties.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Check out the DataFrame: 
melt_properties.dtypes
```

```python
# Make Average_price column float
melt_properties['Average_price'] = pd.to_numeric(melt_properties['Average_price'])
melt_properties.dtypes
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# How many Nan do we have
melt_properties[melt_properties['ID'].isna()]
```

```python
# Look at counts of values
melt_properties.count()
```

```python
# We have 936 rows with NaN to be removed
# set up new dataframe that is free of NaN
all_values_properties = melt_properties.dropna()
all_values_properties.count()
```

```python
all_values_properties['London_Borough'].unique()
```

```python
# Identify the non Borough labels
nonBoroughs = ['Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
              'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
              'SOUTH WEST', 'England']
```

```python
# remove rows from dataframe that are in the nonBoroughs list
df = all_values_properties[~all_values_properties.London_Borough.isin(nonBoroughs)]
df.shape
```

```python
# inspect the final df for values
df.count()
```

```python
df.head()
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# Apply lambda function to add year column to dataframe
df['Year'] = df['Month'].apply(lambda t: t.year)

# Look at the head of the data again to see new column
df.head()
```

```python
# Use 'groupby' to calculate the mean for each year and for each Borough.
# set new dataframe with the grouped data
## As you can see, the variables Borough and Year are now indices
dfg = df.groupby(by=['London_Borough', 'Year']).mean()
dfg.sample(10)
```

```python
# note that the data needs to be reindexed for order
dfg = dfg.reset_index()
# now dates and data should be in ascending order
dfg.head(10)
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# create method for current_price_ratio
# for each borough the ratio of 1998/2018 is returned
def create_price_ratio(d):
    y1998 = float(d['Average_price'][d['Year']==1998])
    y2018 = float(d['Average_price'][d['Year']==2018])
    ratio = [y1998/y2018]
    return ratio
```

```python
#  Test out function by calling it 
#  save into dictionary called final
final = {}
# iterate through each of the unique elements of the 'London_Borough' column
# Call the iterator variable 'b' 
for b in dfg['London_Borough'].unique():
    # Let's make our parameter to our create_price_ratio function 
    borough = dfg[dfg['London_Borough'] == b]
    # Make a new entry in the dictionary based on ratio of borough
    final[b] = create_price_ratio(borough)
# final is a dictionary, make it into a dataframe
df_ratios = pd.DataFrame(final)
df_ratios.head()
```

```python
# All we need to do now is transpose it, and reset the index! 
df_ratios_T = df_ratios.transpose()
df_ratios = df_ratios_T.reset_index()
df_ratios.head()
```

```python
# Let's just rename the 'index' column as 'London_Borough', and the '0' column to '2018'.
df_ratios.rename(columns={'index':'Borough', 0:'2018'}, inplace=True)
df_ratios.head()
```

```python
# Sort the top 15 boroughs in descending order
# assign variable top15 to the result 
top15 = df_ratios.sort_values(by='2018',ascending=False).head(15)
print(top15)
```

```python
# plot the results as a bar graph of top 15 value changes sort descending
vals = top15[['Borough','2018']].plot(kind='bar')

vals.set_xticklabels(top15.Borough)
```

```python
# compare results with the bottom 15 for difference
bottom15 = df_ratios.sort_values(by='2018',ascending=True).head(15)
print(bottom15)
```
